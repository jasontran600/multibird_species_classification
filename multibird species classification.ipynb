### Denoising audio clips with [wavelet packet decomposition and Shannon entropy](https://journals.plos.org/plosone/article?id=10.1371/journal.pone.0146790)
To summarise the approach to bird bioacoustic denoising, I took the following steps:
1. Find a suitable mother wavelet.
2. Find the most suitable decomposition level based on the Shannon entropy.
3. Apply the wavelet transform to the noisy signal to produce the noisy wavelet coefficients.
4. Determine the appropriate threshold to best remove the noise based on the Shannon entropy.
5. Invert the wavelet transform of the retained wavelet coefficients to obtain the denoised signal.
6. Apply a suitable ordinary band-pass or low-pass filter where possible to remove any noise left outside the frequnecy range of the signal.

#### 2. Find the most suitable decomposition level based on its Shannon entropy
Shannon entropy for a wavelet is calculated as:
\begin{equation*}
\ S    = -\sum \limits_{i} s_{i}^2 ln(s_{i}^2)
\end{equation*}
where $s_{i}$ represents the coefficients of signal $s$ in an orthonormal basis. If the entropy value is greater than one, the component has a potential to reveal more information about the signal [1].

Therefore, [Priyadarshani, N. et al](https://journals.plos.org/plosone/article?id=10.1371/journal.pone.0146790) showed that the idea of using entropy for wavelets is to argue that when the entropy is small, the accuracy of the selected wavelet basis is higher, This computation is calculated at each node to choose whether or not to retain a node, and stopped created the tree at the point where all of hte nodes contained noise are removed by thos computation, meaning that the signal was fully described.

Decomposition is continued until the maximum entropy of a parent node (at level $L$) is lower than the maximum entropy of its child nodes (at level $L+1$). At that point the decomposition was stopped, and the best decomposition level is determined as $L$.

* [1] Wang, D., Miao, D., and Xie, C., 2011, Best basis-based wavelet packet entropy feature extraction and hierarchical EEG classification for epileptic detection, Expert Systems with Applications.

```python
# Function for wavelet Shannon entropy
def wavelet_entropy(node_data):
    p = node_data**2/np.sum(node_data**2)
    s = -np.sum(p*np.log(p))
    return s
```

```python
# Collect nodes at specified level
nodes = [node.path for node in wp.get_level(11, 'natural')]
print('Number of nodes: {}'.format(len(nodes)))
print('')

# Calculate Shannon entropy for each node at specified level
S = []
for node in nodes:
    node_data = wp[node].data
    entropy = wavelet_entropy(node_data)
    S.append(entropy)
```
